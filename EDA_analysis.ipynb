```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
df=df.drop("Unnamed: 0",axis=1)
df
```

```python
# renaming the column
d={'cp':'chest_pain_type','trestbps':'resting_bp','chol':'cholestoral','fbs':'fasting_blood_sugar',
   'thalach':'maximum_heart_rate','exang':'exercise_induced_angina','ca':'number of major vessels'
   }
df.rename(columns=d,inplace=True)
df.head()
```

```python
sns.countplot(x="target",data=df)
```

MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df.drop("target",axis=1)
y=df.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=[LogisticRegression(),SVC(kernel="linear"),RandomForestClassifier(n_estimators=100,max_depth=5)]

for m in model:
    model=m
    model=model.fit(X_train,y_train)
    print("MODEL : ",m)
    print("Training efficiency: ",model.score(X_train,y_train))
    print("Testing efficiency: ",model.score(X_test,y_test))
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
print("Training efficiency: ",model.score(X_train,y_train))
print("Testing efficiency: ",model.score(X_test,y_test))
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
print("ACCURACY  : ",accuracy_score(y_test,y_pred))
print("RECALL  : ",recall_score(y_test,y_pred))
```
